```python
import numpy as np
import pandas as pd
```

```python
def warn(*args, **kwargs):
    pass

import warnings

warnings.warn = warn
warnings.filterwarnings("ignore")
```

```python
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

```python
# elimizde bir url var bu url i ilk olarak okumamız gerek 
# tüm sayfayı okuduk 
tables = pd.read_html(URL)
tables
#df.columns = range(df.shape[1])
```

```python
# utl içinde farklı kısımlarda verileri birer eleman oldu. 4. eleman ile çalışacağız 

df = tables[3]
df
```

```python
# sütunların hepsini bir dizi elemanı gibi değiştirdik 
df.columns = range(df.shape[1])
df
```

```python
# ilk iki sütunu aldık 
df = df[[0,2]]
df
```

```python
# ilk 10 değeri çağırdık
df = df.iloc[1:11,:]
df
```

```python
df.columns = ["Country", "GDP (Million USD)"]
df
```

```python
df['GDP (Million USD)'] = df['GDP (Million USD)'].astype(int)
df['GDP (Million USD)']
```

```python
df[["GDP (Million USD)"]] = df [["GDP (Million USD)"]]/1000
df[["GDP (Million USD)"]]
```

```python
df[["GDP (Million USD)"]] = np.round(df[["GDP (Million USD)"]], 2)
df[["GDP (Million USD)"]]
```

```python
df.rename(columns = {"GDP (Million USD)" : "GDP (Billion USD)"})
```

```python
df.to_csv("./Largest_economies.csv")
```
